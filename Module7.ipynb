```python
#-------------------------------------------
# Program Name: Course Information Lookup
# Author: Stephanie Cogburn
# Date: 01/09/2024
#-------------------------------------------
# Pseudocode:
#   1. Create a dictionary for course numbers and their corresponding room numbers.
#   2. Create a dictionary for course numbers and their corresponding instructors.
#   3. Create a dictionary for course numbers and their corresponding meeting times.
#   4. Prompt the user to enter a course number.
#   5. Retrieve and display the entered course's room number, instructor, and meeting time.
#   6. Handle cases where the course number does not exist in the dictionaries.
#-------------------------------------------
# Program Inputs: course_number
# Program Outputs: room_number, instructor, meeting_time
#-------------------------------------------

# Step 1: Define the dictionaries
room_numbers = {
    'CSC101': '3004',
    'CSC102': '4501',
    'CSC103': '6755',
    'NET110': '1244',
    'COM241': '1411'
}

instructors = {
    'CSC101': 'Haynes',
    'CSC102': 'Alvarado',
    'CSC103': 'Rich',
    'NET110': 'Burke',
    'COM241': 'Lee'
}

meeting_times = {
    'CSC101': '8:00 a.m.',
    'CSC102': '9:00 a.m.',
    'CSC103': '10:00 a.m.',
    'NET110': '11:00 a.m.',
    'COM241': '1:00 p.m.'
}

# Step 2: Prompt the user to enter a course number
course_number = input("Enter a course number (e.g., CSC101): ")

# Step 3: Retrieve and display the course information
if course_number in room_numbers:
    print(f"Room Number: {room_numbers[course_number]}")
    print(f"Instructor: {instructors[course_number]}")
    print(f"Meeting Time: {meeting_times[course_number]}")
else:
    print("Course not found. Please enter a valid course number.")
```
